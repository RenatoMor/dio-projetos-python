## *ANÁLISES*

```python
# Exibir dados numéricos para análise------------------------------
colunas_numericas = ['Preco', 'Desconto', 'Vlr Bruto']
descricao = df[colunas_numericas].describe().round(2)
descricao
```

```python
# Exibe dados numéricos para análise por Estado--------------------
colunas_numericas = ['Preco', 'Desconto', 'Vlr Bruto']  # Colunas numéricas
descricao_por_estado = df.groupby('Estado')[colunas_numericas].describe().round(2) # Agrupar Estados calcula estatísticas descritivas

descricao_por_estado_transposta = descricao_por_estado.T  # Facilitar a visualização

descricao_por_estado_transposta
```

```python
# Receita total por Estado ----------------------------------------
receita_por_estado = df.groupby('Estado')['Vlr Bruto'].sum()

for estado, valor_total in receita_por_estado.items():
    
    print(f"\nValor total da receita R$ {valor_total:.2f}: {estado}")
```

```python
# Receita total e top 3 vendedores por estado
vendas_por_vendedor_estado = df.groupby(['Estado', 'Vendedor'])['Vlr Bruto'].sum()
tres_melhores_vendedores_por_estado = vendas_por_vendedor_estado.groupby('Estado').nlargest(3) # Top 3 vendedores 
receita_por_estado = df.groupby('Estado')['Vlr Bruto'].sum()  # Calcular a receita total por estado

# Exibir a receita total por estado em formato de DataFrame
receita_por_estado_df = receita_por_estado.reset_index()
receita_por_estado_df.columns = ['Estado', 'Valor total da receita']
display(receita_por_estado_df)

# Exibir o top 3 vendedores por estado com o percentual de vendas
for estado, valor_total in receita_por_estado.items():
    print(f"\n{estado} - Valor total da receita R$ {valor_total:.2f}")
    
    tres_melhores_vendedores = tres_melhores_vendedores_por_estado.loc[estado] # valores de vendas dos top 3 
    
    for vendedor, valor_vendas in tres_melhores_vendedores.items():
        percentual_vendas = (valor_vendas / valor_total) * 100
        print(f" - {vendedor}: {percentual_vendas:.2f}%")
```

```python
# Receita Anual ---------------------------------------------------
receita_anual_por_estado = df.groupby(['Ano', 'Estado'])['Vlr Bruto'].sum()
print('\nReceita Anual por estado') 
# Exibir os resultados anuais por estado
receita_anual_por_estado  # Exibir os resultados anuais por estado
```

```python
# Exibe as estatísticas descritivas do DataFrame
colunas_numericas = ['Preco', 'Desconto', 'Vlr Bruto']

media = df[colunas_numericas].mean()
mediana = df[colunas_numericas].median()

# Encontrar a moda para cada coluna numérica
moda = pd.Series({coluna: df[coluna].value_counts().idxmax() for coluna in colunas_numericas})

minimo = df[colunas_numericas].min()
maximo = df[colunas_numericas].max()
desvio_padrao = df[colunas_numericas].std()
variancia = df[colunas_numericas].var()

# Criar um DataFrame para cada estatística
media_df = pd.DataFrame(media, columns=['Média'])
mediana_df = pd.DataFrame(mediana, columns=['Mediana'])
moda_df = pd.DataFrame(moda, columns=['Moda'])
minimo_df = pd.DataFrame(minimo, columns=['Mínimo'])
maximo_df = pd.DataFrame(maximo, columns=['Máximo'])
desvio_padrao_df = pd.DataFrame(desvio_padrao, columns=['Desvio Padrão'])
variancia_df = pd.DataFrame(variancia, columns=['Variância'])

# Exibir cada estatística usando display()
display(media_df)
display(mediana_df)
display(moda_df)
display(minimo_df)
display(maximo_df)
display(desvio_padrao_df)
display(variancia_df)
```

```python
# Receita por forma de pagamento----------------------------------------------

# Filtrar as vendas por forma de pagamento
vendas_pix = df[df['Forma de Pgto'] == 'Pix']
vendas_cartao = df[df['Forma de Pgto'] == 'Cartão']
vendas_boleto = df[df['Forma de Pgto'] == 'Boleto']

# Calcular o total de vendas para cada forma de pagamento
total_pix = vendas_pix['Vlr Bruto'].sum()
total_cartao = vendas_cartao['Vlr Bruto'].sum()
total_boleto = vendas_boleto['Vlr Bruto'].sum()

# Exibir os totais de cada forma de pagamento
resultados = pd.DataFrame({
    'Forma de Pgto': ['Pix', 'Cartão', 'Boleto'],
    'Total de vendas': [total_pix, total_cartao, total_boleto]
})

display(resultados)
```

```python
# Grafico de barras horizontal com os 10 produtos mais vendidos---------------------
qtde_produtos_vendidos = df["Produto"].value_counts()

# Selecionar apenas os 10 primeiros resultados
top_10_produtos_vendidos = qtde_produtos_vendidos.head(10)

# Plotar o gráfico de barras horizontal
plt.figure(figsize=(6, 4))  # Tamanho da figura
plt.barh(top_10_produtos_vendidos.index, top_10_produtos_vendidos.values, color='DarkSlateBlue')
plt.xlabel('Quantidade de Vendas')
plt.ylabel('Produto')
plt.title('Top 10 Produtos Mais Vendidos')
plt.gca().invert_yaxis()  # Inverter o eixo y para exibir em ordem decrescente
plt.show()
```

```python
# Converter a coluna de Data para o tipo datetime
df["Data"] = pd.to_datetime(df["Data"])

# Agrupar a receita total por ano
receita_por_ano = df.groupby(df["Data"].dt.year)["Vlr Bruto"].sum()

# Plotar o gráfico de pizza
plt.figure(figsize=(5, 5))
plt.pie(receita_por_ano, labels=receita_por_ano.index, autopct="%1.1f%%", startangle=90)
plt.title('Receita Total por Ano (%)\n\n')
plt.axis('equal')  # Para exibir o gráfico como um círculo
plt.legend(receita_por_ano.index, title="Ano", loc="best", bbox_to_anchor=(1, 0.7))  # Adicionar legenda à direita
plt.style.use('fivethirtyeight')
plt.show()
```

```python
# Converter a coluna "Data" para o tipo datetime caso necessário
df["Data"] = pd.to_datetime(df["Data"])

# Filtrar apenas os dados de 2019
df_2019 = df[df["Data"].dt.year == 2019]

# Agrupar as vendas mensais por estado no ano de 2019
vendas_mensais_por_estado = df_2019.groupby([df_2019["Data"].dt.month, "Estado"])["Vlr Bruto"].sum()

# Resetar o índice para ter acesso aos valores das colunas
vendas_mensais_por_estado = vendas_mensais_por_estado.reset_index()

# Criar uma coluna com o nome do mês
meses = {1: "Jan", 2: "Fev", 3: "Mar", 4: "Abr", 5: "Mai", 6: "Jun", 7: "Jul", 8: "Ago", 9: "Set", 10: "Out", 11: "Nov", 12: "Dez"}
vendas_mensais_por_estado["Mês"] = vendas_mensais_por_estado["Data"].map(meses)

# Plotar o gráfico de vendas mensais de cada estado em 2019
plt.figure(figsize=(12, 4))
for estado in vendas_mensais_por_estado["Estado"].unique():
    vendas_estado = vendas_mensais_por_estado[vendas_mensais_por_estado["Estado"] == estado]
    plt.plot(vendas_estado["Mês"], vendas_estado["Vlr Bruto"], marker='s', label=estado)

plt.xlabel("Mês")
plt.ylabel("Vendas Mensais")
plt.title("Vendas Mensais de cada Estado em 2019")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.xticks(rotation=45)
plt.tight_layout()

# Estilizar o fundo do gráfico
plt.style.use('fivethirtyeight')   # Ou 'dark_background  seaborn-v0_8-colorblind' para fundo escuro

# Exibir o gráfico
plt.show()


```

```python
# Filtrar os dados para considerar apenas as vendas em 2018
df_2018 = df[df['Ano'] == 2018]

# Agrupar os valores por mês e estado para calcular a receita mensal
receita_mensal_por_estado = df_2018.groupby(['Ano', 'Mes', 'Estado'])['Vlr Bruto'].sum().reset_index()

# Obter a lista de estados únicos
estados = receita_mensal_por_estado['Estado'].unique()

# Criar um gráfico para cada estado
for estado in estados:
    df_estado = receita_mensal_por_estado[receita_mensal_por_estado['Estado'] == estado]
    
    plt.figure(figsize=(7, 4))
    plt.bar(df_estado['Mes'], df_estado['Vlr Bruto'], color='blue, gray'.split(', '))
    
    # Configurações adicionais
    plt.xlabel("Mês")
    plt.ylabel("Receita")
    plt.title(f"Receita Mensal em 2018 - {estado}")
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)    
    plt.style.use('fivethirtyeight')      
    # Exibir o gráfico
    plt.tight_layout()
    plt.show()
```
